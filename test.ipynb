```python
import h5py
import numpy as np
import tensorflow as tf

def load_dataset(filename):
    data = np.load(filename)
    
    eeg_data = {}
    for subject_key in set([key.split('_')[0] for key in data.keys()]):
        eeg_data[subject_key] = {}
        for dataset_type in ['train_ds', 'test_ds']:
            x_data = data[f'{subject_key}_{dataset_type}_x']
            y_data = data[f'{subject_key}_{dataset_type}_y']
            x_tensor = tf.convert_to_tensor(x_data, dtype=tf.float32)  # Convert NumPy back to Tensor
            y_tensor = tf.convert_to_tensor(y_data, dtype=tf.float32)  # Convert NumPy back to Tensor
            dataset = tf.data.Dataset.from_tensor_slices((x_tensor, y_tensor)).batch(32)
            eeg_data[subject_key][dataset_type] = dataset
    return eeg_data

dataset_file = 'loading_datasets/bciciv2a_data.npz'
eeg_data = load_dataset(dataset_file)
```
